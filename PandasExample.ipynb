```python
import pandas as pd
```

```python
data_file = 'pattern.txt'

pattern = pd.read_csv(
    data_file,
    sep='\\s+',
    skiprows=2,
    names=['th', 'ph', 'abs', 'abs_th', 'arg_th', 'abs_ph', 'arg_ph', 'ax']
)

# pattern.head(15)
pattern.tail(3)
```

```python
pattern_ph15 = pattern[pattern.ph == 90.0]

pattern_ph15[['th', 'abs']].to_csv('test.csv', index=False)

pattern_ph15.plot(x='th', y='abs')
```

```python
DN = pd.concat([pattern[pattern.ph == 0.0], pattern[pattern.ph == 180.0]])

DN.plot(x='th', y='abs')
```
